## Query Webhose for a set of 100 posts/feeds

```python
# be careful how many times you make this call
# get the first batch
output = webhoseio.query("filterWebContent", query_params)
```

```python
#number of articles, titles and publish dates of the first 100 articles
feeds = []
for item in output['posts']:
    feeds.append(item)
    print(item['title'], item['published'])

print(len(feeds))
```

## Crawl additional batches of articles 

```python
count = 4
while count > 0:    
    output = webhoseio.get_next()
    for item in output['posts']:
        feeds.append(item)
    count -= 1
print(len(feeds))
```

## Store data in JSON format, make sure that the file is readable again back into your program

```python
import json
with open("airbnb_webhose.json", "w") as myfile:
    for feed in feeds:
        line = json.dumps(feed)
        myfile.write(line)
        myfile.write("\n")
```

```python
json_data=open("airbnb_webhose.json").readlines()
feeds_read_from_file = []
for line in json_data:
    feeds_read_from_file.append(json.loads(line))
print(len(feeds_read_from_file))
```

```python
for feed in feeds_read_from_file:
    print(feed['title'],feed['url'])
```
