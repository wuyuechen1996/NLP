# NLP-Assignment3 Basic Text Processing

```python
#Use BeautifulSoup (Links to an external site.) or another HTML parsing package to extract text from the article.
import bs4
from bs4 import BeautifulSoup
from bs4.element import Comment
```

```python
def tag_visible(element):
    if element.parent.name in ['style', 'script', 'head', 'title', 'meta', '[document]']:
        return False
    if isinstance(element, Comment):
        return False
    return True
```

```python
def text_from_html(body):
    soup = BeautifulSoup(body, 'html.parser')
    texts = soup.findAll(text=True)
    visible_texts = filter(tag_visible, texts)  
    return u" ".join(t.strip() for t in visible_texts)
```

```python
import urllib.request
html = urllib.request.urlopen('https://www.cnbc.com/2019/01/17/netflix-price-hike-helps-disney-upcoming-streaming-service-analyst.html').read()
text = text_from_html(html)
print(text)
```

## Use re (regular expression) package to:

```python
#Find all matches of $ amounts in the article
import re
print('$ amounts:')
print(re.findall('\$\d*\.?\d+?', text))

# the position of $ amounts in the article
first_pos=0
ans=[]
for i in range(text.count('$')):
    new_list = text[first_pos:]
    next_pos = new_list.index('$') + 1
    ans.append(first_pos + new_list.index('$'))
    first_pos += next_pos
print('positions: ')
print(ans)
```

```python
#Substitute all numbers with # character and print the output
print(re.sub(r'[0-9]','#',text))
```

```python
#Count (using regular expressions) ”Netflix” and “Disney” mentions 
print('Netflix: '+str(len(re.findall('Netflix', text, flags=0))))
print('Disney: '+str(len(re.findall('Disney', text, flags=0))))
```

## Use NTLK and/or Spacy tokenization features to:

```python
#Tokenize sentences and words
import nltk
from nltk import word_tokenize, sent_tokenize, ngrams, pos_tag, RegexpParser
from collections import Counter
```

```python
sentences = sent_tokenize(text)
for sentence in sentences:
    print(sentence)
    print('\n')
```

```python
tokens = word_tokenize(text)
print(Counter(tokens))
```

```python
#Remove all English stop words
from nltk.corpus import stopwords 

stop_words = set(stopwords.words('english')) 
filtered_sentence = [] 

for t in tokens: 
    if t not in stop_words: 
        filtered_sentence.append(t) 
        
print(filtered_sentence) 
```

```python
#List and count n-grams for any given input n
def ListAndCount(tokens,n):
    for item in ngrams(tokens,n):
        print(item)
    print(Counter(ngrams(tokens,n)))

# example: n=3
ListAndCount(tokens,3)
```

```python
#Lemmatize and deduplicate unigrams into a vocabulary of terms.
from nltk.stem import WordNetLemmatizer

vocabulary = set()

for token in tokens:
    wordnet_lemmatizer = WordNetLemmatizer()
    term = wordnet_lemmatizer.lemmatize(token)
    if term not in vocabulary:
        vocabulary.add(term)

print(vocabulary)
```

```python
#Print bigrams and trigrams in the first 5 sentences
print('Bigrams:')
for sentence in sentences[0:5]:
    tokens = word_tokenize(sentence)
    for item in ngrams(tokens,2):
        print(item)
```

```python
print('Trigrams:')
for sentence in sentences[0:5]:
    tokens = word_tokenize(sentence)
    for item in ngrams(tokens,3):
        print(item)
```

```python
#Print POS tags in the first 5 sentences
for sentence in sentences[0:5]:
    tokens = word_tokenize(sentence)
    print(pos_tag(tokens))
```
